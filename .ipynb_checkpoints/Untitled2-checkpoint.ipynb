```python
X = pd.read_csv('./df_sharpe.csv')
y = pd.read_csv('./y_sharpe.csv')
id_group = X['IDgroup']
X = X.drop(['ID', 'IDgroup'], axis=1)
gkf = GroupKFold(5)
splits = gkf.split(X.loc[:, :], y['smoothed'], id_group)
y = y['smoothed']
```

```python
rf_models = []
train_res = []
test_res = []

gkf = GroupKFold(5)
for i, (train_index, test_index) in enumerate(gkf.split(X.loc[:, :], y, id_group)):
        train_X_fold = X.iloc[train_index, :].loc[:, cols].values
        train_Y_fold = y.iloc[train_index].values
        test_X_fold = X.iloc[test_index, :].loc[:, cols].values
        test_Y_fold = y.iloc[test_index].values
        
        rf = RandomForestRegressor(criterion='mae', max_depth=2, n_jobs=10)
        rf.fit(train_X_fold, train_Y_fold)
        
        train_preds = rf.predict(train_X_fold)
        test_preds = rf.predict(test_X_fold)
        mae_train = mean_absolute_error(train_Y_fold, train_preds)
        mae_test = mean_absolute_error(test_Y_fold, test_preds)
        train_res.append(mae_train)
        test_res.append(mae_test)
        
        rf_models.append(rf)
        
np.mean(train_res), np.mean(test_res)
```
